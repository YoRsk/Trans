```python
import ee
import folium

# 初始化Google Earth Engine
ee.Initialize()

# 定义时间范围
start_date = '2010-01-01'
end_date = '2010-12-31'

# 定义地理范围（使用坐标定义一个矩形）
geometry = ee.Geometry.Rectangle([103.843, 1.309, 104.041, 1.456])

# 导入EO-1 Hyperion数据
hyperion = ee.ImageCollection('EO1/HYPERION')

# 过滤图像集合
filtered = hyperion.filterDate(start_date, end_date).filterBounds(geometry)

# 计算过滤后图像集合的中位数
median_image = filtered.median()

# 定义显示参数
visualization_params = {
    'bands': ['B29', 'B20', 'B11'],
    'min': 0.0,
    'max': 3000.0,
    'gamma': 1.0
}

# 将地图添加到Folium
def add_ee_layer(self, ee_image_object, vis_params, name):
    map_id_dict = ee.Image(ee_image_object).getMapId(vis_params)
    folium.raster_layers.TileLayer(
        tiles=map_id_dict['tile_fetcher'].url_format,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        name=name,
        overlay=True,
        control=True
    ).add_to(self)

folium.Map.add_ee_layer = add_ee_layer

# 创建地图对象并设置中心
map_obj = folium.Map(location=[1.38, 103.94], zoom_start=11)

# 将EO-1 Hyperion数据添加到地图
map_obj.add_ee_layer(median_image, visualization_params, 'EO-1 Hyperion')

# 添加图层控制器
map_obj.add_child(folium.LayerControl())

# 显示地图
map_obj
```
